```python
from Bio import SeqIO

# создание хэш-таблицы
sars_cov2_seq = list(SeqIO.parse('SARS-CoV-2.fasta', 'fasta'))[0].seq
k = int(input())
k_dict = dict()

for i in range(len(sars_cov2_seq) - k + 1):
    k_mer = str(sars_cov2_seq[i:i+k])
    if k_mer not in k_dict:
        k_dict[k_mer] = [i]
    else:
        k_dict[k_mer].append(i)
k_dict
```

```python
# поиск k
k = 1
while True:
    k_dict = dict()
    for i in range(len(sars_cov2_seq) - k + 1):
        k_mer = str(sars_cov2_seq[i:i+k])
        if k_mer not in k_dict:
            k_dict[k_mer] = [i]
        elif k_mer in k_dict:
            k += 1
            break
    if len(k_mer) < k:  # это условие проверяет вышли ли мы из цикла, из-за того, что k-мер
        # встретился в словаре, или из-за того, что цикл просто закончился, т.к. прошел без входа в elif
        # таким образом k мы печатаем только если цикл заврершился полностью, т.е каждый k-мер уникален
        continue
    print(k)
    break
```
