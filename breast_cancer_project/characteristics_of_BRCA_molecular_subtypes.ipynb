```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("BRCA_pam50.tsv", sep="\t", index_col=0)
df
```

```python
sns.set(rc={'figure.figsize':(8, 5)})
sns.boxplot(data=df, y=df['ESR1'], x='Subtype');
```

```python
sns.set(rc={'figure.figsize':(8, 5)})
sns.boxplot(data=df, y=df['PGR'], x='Subtype');
```

```python
sns.set(rc={'figure.figsize':(8, 5)})
sns.boxplot(data=df, y=df['ERBB2'], x='Subtype');
```

```python
sns.set(rc={'figure.figsize':(8, 5)})
sns.boxplot(data=df, y=df['MKI67'], x='Subtype');
```

```python
X = df.iloc[:, :-1].to_numpy()
model = PCA()
model.fit(X)
X_pca = model.transform(X)
df.iloc[:, :-1] = X_pca
df.columns = [f"PC{i + 1}" for i in range(X.shape[1])] + ["Subtype"]
sns.scatterplot(x="PC1", y="PC2", hue="Subtype", data=df);
```

```python
indexes = ['ESR1', 'PGR', 'ERBB2', 'MKI67']
d = {'Normal-like': ['средняя экспрессия']*4,
    'Luminal A': ['высокая экспрессия', 'средняя экспрессия', 'средняя экспрессия', 'средняя экспрессия'] ,
    'Luminal B ': ['высокая экспрессия', 'средняя экспрессия', 'средняя экспрессия', 'высокая экспрессия'],
    'Triple-negative': ['нет экспрессии', 'нет экспрессии', 'низкая экспрессия', 'высокая экспрессия'],
    'HER2-enriched': ['низкая экспрессия', 'нет экспрессии', 'высокая экспрессия', 'высокая экспрессия'],
    'Healthy': ['средняя экспрессия']*3 + ['низкая экспрессия']}
pd.DataFrame(data=d, index=indexes)
```

Сравнивая с графиком первых двух компонент видим, что Healthy, Normal-like и Triple-negative хорошо отделяются от Luminal A, Luminal B и HER2-enriched по первой главной компоненте. Действительно, они отличаются по экспрессии

* Сравнивая с графиком первых двух компонент видим, что Luminal A и Luminal B достаточно похожи и даже пересекаются между собой. Действительно, по профилю экпресии данных генов они похожи, отличие наблюдается в экпрессии MKI67.
* Также Luminal B достаточно близок и пересекается на графике с  HER2-enriched. Они близки по степени экспресии MKI67, с другими генами наблюдаются различия.
* Между Healthy и Normal-like различие наблюдается в экпрессии MKI67, остальные гены синтезируются примерно одинаково. В координатах первых двух компонент мы также видим, что эти группы находятся близко. 
* Triple-negative достаточно сильно отделена от других групп на графике, практически без пересечений: у этого типа рака не экпрессируются 2 гена и 1 экспрессирован слабо, в то время как у других типов рака они экпрессируются (кроме  HER2-enriched, который также не экспрессирует PGR)
