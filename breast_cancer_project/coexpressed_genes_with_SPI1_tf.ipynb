```python
import pandas as pd
from scipy.stats import *

df = pd.read_csv('healthy_breast.tsv', sep="\t", index_col=0)

df = df.T
SPI1 = df['SPI1']

coexp_genes = [elem for elem in df if abs(spearmanr(SPI1, df[elem])[0]) > 0.8
               and spearmanr(SPI1, df[elem])[1] < 0.05]

print('\n'.join(coexp_genes))

# копируем список генов и смотрим, что покажет DAVID
# c p-value, скорректированным на множественность по Benjamini, значительно меньшем, чем
# 0.05 показывает, что эти коэкспрессированные гены участвуют в иммунных ответах,
# таким образом можем предположить, что этот ТФ вовлечен в регуляцию иммунитета
```
